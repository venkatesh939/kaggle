```python
data=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv')
data
```

```python
data.drop(['Country name','Regional indicator'],axis=1,inplace=True)
data
```

```python
y=data['Ladder score']
x=data.drop('Ladder score',axis=1)
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
print("Linear regression :",model.score(x_test,y_test))
```
